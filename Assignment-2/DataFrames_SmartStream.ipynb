In this question, you will create a smoothed engagement metric based on users’ average daily watch time. Real user behavior data is often noisy—some days users watch a lot, other days very little. Smoothing helps reveal the underlying trend by reducing short-term fluctuations.

You will implement exponential smoothing manually, using a loop. This is intentional: although Pandas provides built-in tools for smoothing, this task is designed to help you understand how the smoothing process works step by step and to practice a legitimate use case for iteration where each row depends on the previous one.

**What is exponential smoothing?**

Exponential smoothing creates a new value that is a weighted average of:
*	the current observation, and
*	the previous smoothed value

More recent observations are given more weight, while older observations gradually fade in influence.

**What is alpha (α)?**
*	Alpha (α) is the smoothing parameter.
*	It controls how much weight is placed on the current value versus past values.
*	Alpha must be between 0 and 1.

In this problem:
*	α = 0.30, meaning:
*	30% weight is given to the current avg_watch_minutes
*	70% weight is given to the previous smoothed value

A higher alpha reacts more quickly to changes; a lower alpha produces smoother, slower-moving results.

**Why must you use a loop?**

Each smoothed value depends on the previous smoothed value. This creates a row-to-row dependency, which means a vectorized solution is not appropriate here. A loop is required so that each calculation happens in sequence.
